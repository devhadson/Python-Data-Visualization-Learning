# Principales gráficos estadísticos.

Los gráficps incluye su descripción, características clave, casos de uso ideales y el código limpio en Python utilizando [Matplotlib](https://matplotlib.org/) y [Seaborn](https://seaborn.pydata.org/) para su implementación automatizada.

## 1. Gráfico de Barras (Bar Chart)
 
* **Descripción:** Representa datos categóricos con barras rectangulares de longitudes proporcionales a los valores que representan.
* **Características:** Eje x categórico, eje y numérico. Las barras se separan para denotar discontinuidad.
* **Caso de uso:** Comparar magnitudes discretas entre grupos independientes.

    ### Código en Python:

```python
import matplotlib.pyplot as plt
import seaborn as sns

categorias = ['Producto A', 'Producto B', 'Producto C', 'Producto D']
ventas = [350, 480, 290, 610]

#fig = plt.figure(figsize=(8, 5))
#fig.suptitle("Hadson.Tech")
plt.figure(figsize=(8, 5))
sns.barplot(x=categorias, y=ventas, palette='viridis')
plt.title('Comparativa de Ventas por Producto')
plt.xlabel('Categorías')
plt.ylabel('Unidades Vendidas')
plt.show()
```

## 2. Gráfico de Líneas (Line Chart)

* **Descripción:** Conecta puntos de datos individuales con segmentos de línea recta continuos.
* **Características:** Muestra continuidad matemática o temporal. Excelente para identificar pendientes y direcciones.
* **Caso de uso:** Monitorear y proyectar tendencias a lo largo de intervalos de tiempo continuos.

    ### Código en Python:

```python
import matplotlib.pyplot as plt
import numpy as np

meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']
ingresos = [12000, 14500, 13000, 16800, 18200, 21000]

plt.figure(figsize=(8, 5))
plt.plot(meses, ingresos, marker='o', linestyle='-', color='teal', linewidth=2)
plt.title('Evolución de Ingresos Mensuales')
plt.xlabel('Meses')
plt.ylabel('Monto ($)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

## 3. Histograma (Histogram)

* **Descripción:** Agrupa datos numéricos continuos en contenedores (bins) y muestra la frecuencia de observaciones en cada intervalo.
* Características: Las barras están totalmente unidas para reflejar la continuidad de la variable. Revela la forma de la distribución.
* **Caso de uso:** Evaluar la simetría, sesgo y dispersión de variables numéricas puras.

    ### Código en Python:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generar 1000 estaturas simuladas (media 1.70, desviación 0.08)
estaturas = np.random.normal(loc=1.70, scale=0.08, size=1000)

plt.figure(figsize=(8, 5))
sns.histplot(estaturas, bins=25, kde=True, color='crimson')
plt.title('Distribución de Estaturas en la Población')
plt.xlabel('Estatura (metros)')
plt.ylabel('Frecuencia Absoluta')
plt.show()
```

## 4. Gráfico Circular / Sectores (Pie Chart)

* **Descripción:** Círculo dividido en porciones geométricas donde el ángulo central representa una fracción del total.
* **Características:** Representa un total cerrado (100%). Requiere pocas categorías (máximo 5) para mantener la legibilidad.
* **Caso de uso:** Mostrar la composición estática y relativa de un mercado o presupuesto.

    ### Código en Python:

```python
import matplotlib.pyplot as plt

mercado = ['Empresa A', 'Empresa B', 'Empresa C', 'Otros']
cuotas = [40, 30, 15, 15]
colores = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

plt.figure(figsize=(6, 6))
plt.pie(cuotas, labels=mercado, autopct='%1.1f%%', startangle=90, colors=colores, explode=(0.05, 0, 0, 0))
plt.title('Cuota de Mercado Global')
plt.show()
```

## 5. Gráfico de Dispersión (Scatter Plot)

* **Descripción:** Mapea coordenadas (X, Y) individuales basadas en dos propiedades numéricas independientes para cada observación.
* **Características:** Identifica nubes de puntos, patrones geométricos directos, inversos o nulos.
* **Caso de uso:** Analizar la existencia y el tipo de correlación lineal o no lineal entre dos métricas continuas.

    ### Código en Python:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

horas_estudio = np.random.uniform(1, 20, 100)
calificaciones = 40 + 3 * horas_estudio + np.random.normal(0, 5, 100)

plt.figure(figsize=(8, 5))
sns.scatterplot(x=horas_estudio, y=calificaciones, color='purple', alpha=0.8)
plt.title('Relación entre Horas de Estudio y Calificación')
plt.xlabel('Horas Semanales Dedicadas')
plt.ylabel('Puntaje Obtenido')
plt.grid(True, alpha=0.3)
plt.show()
```

## 6. Diagrama de Caja y Bigotes (Box-Plot)

* **Descripción:** Resume gráficamente una distribución numérica mediante sus cinco estadísticos numéricos esenciales: mínimo, Q1, mediana, Q3 y máximo.
* **Características:** Destaca visualmente la dispersión intercuartílica y aísla de forma explícita los valores atípicos (outliers).
* **Caso de uso:** Comparar la dispersión, simetría y variabilidad interna de múltiples muestras numéricas.

    ### Código en Python:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Datos simulados de salarios en 3 industrias
data = pd.DataFrame({
    'Salario': np.concatenate([np.random.normal(2500, 400, 100), np.random.normal(3800, 600, 100), np.random.normal(1900, 200, 100)]),
    'Sector': ['Tech']*100 + ['Finanzas']*100 + ['Retail']*100
})

plt.figure(figsize=(8, 5))
sns.boxplot(x='Sector', y='Salario', data=data, palette='Set2')
plt.title('Distribución de Salarios Mensuales por Sector')
plt.xlabel('Sector Económico')
plt.ylabel('Salario ($)')
plt.show()
```

## 7. Gráfico de Densidad (KDE Plot)

* **Descripción:** Versión suavizada y no paramétrica del histograma basada en la estimación de la densidad de kernel.
* **Características:** Reemplaza los bloques discretos por curvas continuas de probabilidad acumulada uniforme.
* **Caso de uso:** Visualizar la forma pura de la distribución de una variable ignorando las divisiones del binning artificial.

    ### Código en Python:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

edades = np.random.exponential(scale=30, size=500) + 18

plt.figure(figsize=(8, 5))
sns.kdeplot(edades, fill=True, color='darkorange', alpha=0.4, linewidth=2)
plt.title('Curva Continuada de Densidad de Edades de Usuarios')
plt.xlabel('Edad')
plt.ylabel('Densidad Estructural')
plt.show()
```

## 8. Gráfico de Áreas (Area Chart)

* **Descripción:** Gráfico de líneas donde el espacio geométrico comprendido entre la línea de tendencia y el eje de abscisas se rellena con color.
* **Características:** Acentúa la percepción del volumen o magnitud acumulada total sobre el tiempo.
* **Caso de uso:** Comunicar la evolución de totales agregados compuestos por varias subcategorías acumulativas.

    ### Código en Python:

```python
import matplotlib.pyplot as plt
import pandas as pd

tiempo = range(1, 6)
energia_solar = [10, 15, 22, 35, 50]
energia_eolica = [20, 28, 32, 40, 45]

plt.figure(figsize=(8, 5))
plt.stackplot(tiempo, energia_solar, energia_eolica, labels=['Solar', 'Eólica'], colors=['#ffdd57', '#23d160'], alpha=0.7)
plt.title('Evolución de Generación de Energía Renovable Acumulada')
plt.xlabel('Años de Operación')
plt.ylabel('Megavatios (MW)')
plt.legend(loc='upper left')
plt.show()
```

## 9. Gráfico de Violín (Violin Plot)

* **Descripción:** Fusión directa entre un diagrama de caja convencional y una curva simétrica de estimación de densidad de kernel a ambos lados.
* **Características:** Muestra los estadísticos de cuartiles internos junto con la distribución bimodal o multimodal detallada del dataset.
* **Caso de uso:** Comparar variables continuas complejas cuya densidad interna posee múltiples picos invisibles en un box-plot estándar.

    ### Código en Python:

```python
import matplotlib.pyplot as plt
import seaborn as sns

iris = sns.load_dataset('iris')

plt.figure(figsize=(8, 5))
sns.violinplot(x='species', y='sepal_length', data=iris, palette='muted', inner='quartile')
plt.title('Comparativa de Longitud de Sépalo por Especie')
plt.xlabel('Especie de Planta')
plt.ylabel('Longitud de Sépalo (cm)')
plt.show()
```

## 10. Mapa de Calor (Heatmap)
 
* **Descripción:** Matriz bidimensional donde los valores numéricos individuales de las celdas se sustituyen por una codificación cromática proporcional.
* **Características:** Permite procesar de forma inmediata patrones masivos de correlación cruzada en tablas de datos extensas.
* **Caso de uso:** Visualizar matrices de coeficientes de correlación de Pearson o densidades de ocurrencias cruzadas.

    ### Código en Python:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generar una matriz de correlación aleatoria simétrica
matriz_datos = np.random.rand(5, 5)
correlacion = np.corrcoef(matriz_datos)
variables = ['Var 1', 'Var 2', 'Var 3', 'Var 4', 'Var 5']

plt.figure(figsize=(7, 5))
sns.heatmap(correlacion, annot=True, cmap='coolwarm', xticklabels=variables, yticklabels=variables, vmin=-1, vmax=1)
plt.title('Matriz de Correlaciones Estadísticas')
plt.show()
```

## 11. Gráfico de Burbujas (Bubble Chart)

* **Descripción:** Extensión del gráfico de dispersión tradicional donde los puntos adquieren una tercera y cuarta dimensión mediante su tamaño físico y su color asignado.
* **Características:** Permite representar relaciones multivariadas complejas en solo dos dimensiones geométricas de pantalla.
* **Caso de uso:** Evaluar relaciones socioeconómicas donde se necesita ponderar el volumen poblacional o de inversión por cada observación.

    ### Código en Python:

```python
import matplotlib.pyplot as plt
import numpy as np

pib_per_capita = np.random.uniform(5000, 50000, 30)
esperanza_vida = 60 + (pib_per_capita / 1000) * 0.4 + np.random.normal(0, 2, 30)
poblacion_millones = np.random.uniform(5, 300, 30)

plt.figure(figsize=(9, 6))
burbujas = plt.scatter(pib_per_capita, esperanza_vida, s=poblacion_millones*2, c=esperanza_vida, cmap='plasma', alpha=0.6, edgecolors="grey", linewidth=1)
plt.title('Análisis Macroeconómico de Desarrollo')
plt.xlabel('PIB per Cápita (USD)')
plt.ylabel('Esperanza de Vida (Años)')
plt.colorbar(burbujas, label='Gradiente de Esperanza de Vida')
plt.show()
```

## 12. Gráfico de Coordenadas Paralelas (Parallel Coordinates)

* **Descripción:** Mapea observaciones de datasets con alta dimensionalidad sobre un conjunto de ejes paralelos equidistantes verticales, conectando cada muestra con una línea quebrada continua.
* **Características:** Rompe la restricción geométrica de las tres dimensiones clásicas del plano cartesiano convencional.
* **Caso de uso:** Clasificación multivariable y detección rápida de clusters o agrupamientos en minería de datos avanzada.

    ### Código en Python:

```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import parallel_coordinates
import seaborn as sns

iris = sns.load_dataset('iris')

plt.figure(figsize=(10, 5))
parallel_coordinates(iris, 'species', color=('#1f77b4', '#ff7f0e', '#2ca02c'), alpha=0.5)
plt.title('Clasificación Multivariada de Iris mediante Coordenadas Paralelas')
plt.xlabel('Atributos de la Planta')
plt.ylabel('Valor de Medición (cm)')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.show()
```
