# Principales Características de un gráfico

Un gráfico estadístico o de datos es una representación visual que transforma números en información fácil de entender. Para que sea efectivo y profesional, debe contar con las siguientes **características esenciales**:

## Estructura y Elementos Clave

* **Título claro:** Describe qué mide el gráfico y el contexto de los datos.
* **Ejes coordenados:** Línea horizontal (X) para variables y línea vertical (Y) para magnitudes.
* **Etiquetas de eje:** Nombres explícitos en cada eje con sus unidades de medida respectivas.
* **Escala proporcional:** Graduación numérica que mantiene distancias matemáticas exactas entre los valores.
* **Leyenda informativa:** Guía que identifica el significado de diferentes colores, series o símbolos.
* **Líneas de división:** Cuadrícula de fondo opcional que facilita la lectura precisa de puntos.
* **Fuente de datos:** Nota al pie que indica la procedencia u origen de la información.

## Cualidades Técnicas y Visuales

* **Simplicidad visual:** Evita elementos decorativos innecesarios que distraigan al lector del mensaje principal.
* **Legibilidad tipográfica:** Textos con fuentes limpias y tamaños proporcionales al espacio disponible.
* **Contraste cromático:** Uso estratégico del color para diferenciar categorías sin saturar la vista.
* **Fidelidad matemática:** Representación geométrica exacta que impide la manipulación o distorsión de la realidad.
* **Autonomía total:** Capacidad del gráfico para explicarse por sí mismo sin leer el texto del reporte. [8, 9]

## Componentes Estructurales
* **Figure (Figura):** El lienzo o contenedor global que engloba a todo el gráfico, incluyendo los ejes, títulos, leyendas y subgráficos.
* **Axes (Ejes/Gráfico):** El área de dibujo específica donde se grafican los datos. Una figura puede contener uno o varios Axes.
* **Spines (Bordes/Espinas):** Las líneas que delimitan el área del gráfico y conectan las marcas de los ejes. Son los bordes superior, inferior, izquierdo y derecho.

## Textos e Identificadores

* **Title (Título):** El texto principal ubicado en la parte superior que describe el propósito general del gráfico.
* **X axis label (Etiqueta del eje X):** El texto que identifica la variable explicada en el eje horizontal.
* **Y axis label (Etiqueta del eje Y):** El texto que identifica la variable explicada en el eje vertical.
* **Legend (Leyenda):** El cuadro explicativo que vincula los colores o estilos de las líneas/marcadores con sus respectivos nombres de datos ("Blue signal", "Red signal").

## Representación de Datos

* **Line (line plot) (Línea / Gráfico de líneas):** Representación continua que conecta puntos de datos secuenciales (como las curvas azul y roja).
* **Markers (scatter plot) (Marcadores / Gráfico de dispersión):** Los puntos o símbolos individuales que representan coordenadas específicas de datos aislados en el mapa.

## Elementos de Medición y Guía

* **Major tick (Marca mayor):** Las líneas de graduación principales en los bordes de los ejes que indican valores numéricos clave.
* **Minor tick (Marca menor):** Subdivisiones más pequeñas situadas entre las marcas mayores para lecturas más precisas.
* **Major tick label (Etiqueta de marca mayor):** Los números grandes asignados a las marcas principales (ej. 1, 2, 3).
* **Minor tick label (Etiqueta de marca menor):** Los números más pequeños asignados a las subdivisiones (ej. 0.25, 0.50, 0.75).
* **Grid (Cuadrícula / Rejilla):** Las líneas discontinuas de fondo que se extienden desde las marcas de los ejes para facilitar la lectura visual de las coordenadas.

A modo resumen las características y componentes de un gráfico debe ser claro en 2 aspecto, basicamente en **El "Cómo" debe ser**, y **El "Qué" debe tener**:

## Características fundamentales (El "Cómo" debe ser)

* **Autónomo:** Debe explicarse por sí mismo sin necesidad de leer textos adicionales.
* **Preciso:** Representa los datos con total fidelidad matemática, sin distorsionar la realidad.
* **Simple:** Elimina decoraciones innecesarias (como efectos 3D) para enfocarse en los datos.
* **Legible:** Utiliza tipografías claras, tamaños adecuados y colores con buen contraste.
* **Proporcional:** Mantiene escalas constantes y uniformes en toda su extensión.

## Componentes esenciales (El "Qué" debe tener)

* **Título:** Frase superior que define claramente qué mide el gráfico y su contexto.
* **Ejes (X / Y):** Líneas de referencia horizontal (variables/tiempo) y vertical (magnitudes/cantidades).
* **Etiquetas de eje:** Textos que nombran las variables y especifican sus unidades de medida (ej. kg, $, meses).
* **Escala / Graduación:** Marcas numéricas o categóricas que dividen los ejes de forma equidistante.
* **Leyenda:** Recuadro que descodifica el significado de los colores, formas o tipos de líneas.
* **Datos (Cuerpo):** Las líneas, barras, puntos o sectores que representan físicamente la información.
* **Cuadrícula (Grid):** Líneas de fondo opcionales que ayudan a proyectar visualmente los puntos hacia los ejes.
* **Fuente:** Nota al pie que indica el origen de los datos recopilados para dar credibilidad.
